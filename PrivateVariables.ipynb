```python
class Dog:
    def __init__(self, name, breed):
        self.__name = name
        self.__breed = breed
        
    def say_hello(self):
        print(f'Hi, Im {self.__name} and Im a {self.__breed}')
```

```python
hopi = Dog('Hope', 'Beagle')
hopi.__name  = 'asdf'
hopi.say_hello()
```

```python
print(Dog.__dict__)
print(hopi.__dict__)
```
